```python
import numpy as np    
import scipy.linalg as la
import math
import matplotlib.pyplot as pl
```

```python
def gauss(A: np.array, b: np.array):
    """Solves equation Ax=b for invertible matrix A, using gaussian elimination.
    A is numpy array (n, n)
    b is numpy array (n, 1)
    returns x as numpy array (n, 1)"""
    assert(A.shape[0] == A.shape[1] == b.shape[0] and b.shape[1] == 1)
    n = A.shape[0]
    A = np.hstack((A, b))
    
    for i in range(n):
        max_i = i + np.abs(A[i:,i]).argmax()
        A[[i, max_i], :] = A[[max_i, i],:]        
        A[i] /= A[i, i] 
        A[i + 1:] -= np.tile(A[i], (n - i - 1, 1)) * A[i + 1:, i].reshape((n - i - 1, 1))
    x = A[:, n].copy()
    for i in reversed(range(0, n)):
        x[i] -= np.sum(A[i, i + 1:n] * x[i + 1:])
    return x.reshape((n, 1))
```

```python
def cholesky(A: np.array):
    assert(A.shape[0] == A.shape[1])
    n = A.shape[0]
    L = A.copy()
    for i in range(n):
        if i > 0:
            L[i:n, i] = L[i:n, i] - np.dot(L[i:n, :i], L[i, :i].transpose())
        L[i:n, i] = L[i:n, i] / math.sqrt(L[i,i])
        L[:i, i] = 0
    return L
```

```python
n, p = 40, 20
a = np.exp(np.random.rand(n, n) * p / 2)
a = np.dot(a, a.transpose())
np.max(np.abs(cholesky(a) - la.cholesky(a, lower=True)))
```

```python
def solveCholesky(A: np.array, b: np.array):
    assert(A.shape[0] == A.shape[1] == b.shape[0] and b.shape[1] == 1)
    n = A.shape[0]
    L = cholesky(A)
    x = np.empty(n)
    for i in range(0, n):
        x[i] = (b[i, 0] - np.sum(x[:i] * L[i, :i])) / L[i, i]
    L = L.transpose()
    for i in reversed(range(0, n)):
        x[i] = (x[i] - np.sum(L[i, i + 1:] * x[i + 1:])) / L[i, i]
    return x.reshape(n, 1)
```

```python
def solveCholeskyElimination(A: np.array, b: np.array):
    assert(A.shape[0] == A.shape[1] == b.shape[0] and b.shape[1] == 1)
    n = A.shape[0]
    L = cholesky(A)
    A = np.hstack((L, b))
    for i in range(0, n):
        A[i] /= A[i, i]
        A[i + 1:] -= np.tile(A[i], (n - i - 1, 1)) * A[i + 1:, i].reshape((n - i - 1, 1))
    b = A[:, n].reshape(n, 1)
    A = np.hstack((L.transpose(), b))
    for i in reversed(range(0, n)):
        A[i] /= A[i, i]
        A[:i] -= np.tile(A[i], (i, 1)) * A[:i, i].reshape((i, 1))   
    return A[:, n].reshape(n, 1)
```

```python
n, p = 100, 1
a = np.exp(np.random.rand(n, n) * p / 2)
b = np.exp(np.random.rand(n, 1) * p)
print(np.max(a), np.min(a), np.mean(a), np.median(a))
print(np.max(b), np.min(b), np.mean(b), np.median(b))
```

```python
a = np.dot(a, a.transpose())
print(np.max(a), np.min(a), np.mean(a), np.median(a))
```

```python
x = gauss(a, b)
y = np.abs(x)
print(np.max(y), np.min(y), np.mean(y), np.median(y))
err = np.abs(np.dot(a, x) - b)
print(np.max(err), np.min(err), np.mean(err), np.median(err))
```

```python
x = solveCholesky(a, b)
y = np.abs(x)
print(np.max(y), np.min(y), np.mean(y), np.median(y))
err = np.abs(np.dot(a, x) - b)
print(np.max(err), np.min(err), np.mean(err), np.median(err))
```

```python
p = 20
step, limit = 50, 1500
steps = range(step, limit, step)
gaussTiming = []
choleskyTiming = []
matrixSizes = []
for i, n in enumerate(steps):
    a = np.exp(np.random.rand(n, n) * p / 2)
    b = np.exp(np.random.rand(n, 1) * p)
    a = np.dot(a, a.transpose())
    x = %timeit -o gauss(a, b)
    gaussTiming.append(x.best ** (1.0/3))
    x = %timeit -o solveCholesky(a, b)
    choleskyTiming.append(x.best ** (1.0/3))
    matrixSizes.append(n)
    pl.plot(matrixSizes, gaussTiming, 'r--', matrixSizes, choleskyTiming, 'bs')
    pl.show()
```
